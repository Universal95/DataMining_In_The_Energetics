```python
import pandas as pd
import numpy as np
```

```python
weather = pd.read_csv('idojaras_reszletes.csv')
```

```python
new_weather = pd.DataFrame();

for i in range(0, 10):
    new_weather = new_weather.append(weather[weather['time'] == '0'+ str(i) + ':00'])
for i in range(10, 24):
    new_weather = new_weather.append(weather[weather['time'] == str(i) + ':00'])
new_weather = new_weather.sort(['date', 'time'])
    
for row in range(0, len(new_weather)):
    new_weather['time'].iloc[row] = int(new_weather['time'].iloc[row].split(':')[0])
    if (row % 100 == 0):
        print(row)
    
new_weather
```

```python
data = pd.read_csv('Data_Expansion.csv')
new_data = pd.merge(data, new_weather, on='date')

new_data.__delitem__('max')
new_data.__delitem__('min')
new_data.__delitem__('rain')
new_data.__delitem__('snow')
new_data.__delitem__('avg_max')
new_data.__delitem__('avg_min') 
# new_data.__delitem__('time')

new_data
```
