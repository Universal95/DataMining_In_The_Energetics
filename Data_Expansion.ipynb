# Homersekletenkent az atlag es szoras kiszamolasa

```python
database.__delitem__('date')
database.__delitem__('tick')
database.__delitem__('min')
database.__delitem__('rain')
database.__delitem__('snow')
database.__delitem__('avg_max')
database.__delitem__('avg_min')
database.__delitem__('day_of_week')
database.__delitem__('holiday')
database.__delitem__('workday')
```

```python
means = pd.DataFrame()
mean = [] 

for i in range(database['max'].min(), database['max'].max()+1):
    
    mean = np.mean(database[database['max']==i])
    means = means.append(mean, ignore_index=True)

means = means.dropna()
cols = means.columns.tolist()
cols = cols[-2:] + cols[:-2]
means = means[cols]
means = means.rename(index=str, columns={'max': '˚C', 'system_load':'avg_sysload', \
                                         'export': 'avg_export', 'import': 'avg_import'})
# means
```

```python
stds = pd.DataFrame()
std = []

for i in range(database['max'].min(), database['max'].max()+1):
    
    std = np.std(database[database['max']==i], axis=0)
    std['max']=i
    stds = stds.append(std, ignore_index=True)

stds = stds.dropna()
cols = stds.columns.tolist()
cols = cols[-2:] + cols[:-2]
stds = stds[cols]
stds = stds.rename(index=str, columns={'max': '˚C', 'system_load':'std_sysload', \
                                       'export': 'std_export', 'import': 'std_import'})
# stds
```

```python
values = pd.merge(means, stds, on='˚C')
values = values.round(1)
values.to_csv('Means_and_stds_for_temperatures.csv', sep=',', index=False)
values
```

```python
database = pd.read_csv('Data_Expansion.csv')
database = pd.merge(database, values, left_on='max', right_on='˚C')
database.__delitem__('˚C')
database = database.sort(['date', 'tick'])
database.to_csv('Data_Expansion.csv', sep=',', index=False)
database
```
