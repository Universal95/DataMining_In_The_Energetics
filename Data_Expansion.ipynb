```python
import pandas as pd
import math
import matplotlib.pyplot as plt
import matplotlib
from matplotlib.dates import YearLocator, MonthLocator, DateFormatter
from datetime import datetime 
import numpy as np
```

```python
data = pd.read_csv('database_2017_03.csv')
wheater = pd.read_csv('weather_info.csv')
database = pd.merge(data, wheater, on='date')
database['date'] = pd.to_datetime(database['date'])
database['day_of_week'] = np.nan
print(len(database))
for i in range(0, len(database)):
    database.set_value(i, 'day_of_week', database.loc[i]['date'].weekday())
    database.set_value(i, 'tick', database.loc[i]['tick'].split('-')[0])
    if ((i % 1000) == 0):
        print(i)
# database
```
